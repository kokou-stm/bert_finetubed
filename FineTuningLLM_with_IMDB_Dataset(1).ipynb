## Preprocessing (Tokenize dataset  with map)

```python
#Se fait comme une iteration mais en load batch
def preprocess_data(example):
  return tokenizer(example["text"], truncation = True, padding = "max_length", max_length=512)
```

```python
example = dataset['train'][0]
print(example.keys())
print(tokenizer(example["text"], truncation = True, padding = "max_length", max_length=512).keys())
```

```python
tokenized_datasets  = dataset.map(preprocess_data, batched=True) # On peux spécifier aussi le batch_size ( batch_size=16)
```

```python
tokenized_datasets
```

```python
print(tokenized_datasets["train"][0]["text"])
print(tokenized_datasets["train"][0]["label"])
print(tokenized_datasets["train"][0]["input_ids"])
print(tokenized_datasets["train"][0]["token_type_ids"])
print(tokenized_datasets["train"][0]["attention_mask"])
```

- Convert data to the correct format for training

```python
#remove
tokenized_datasets = tokenized_datasets.remove_columns("text")
#rename label column
tokenized_datasets = tokenized_datasets.rename_column('label', "labels")
tokenized_datasets
```

- Convert to Torch format

```python
tokenized_datasets.reset_format()
print("type dataset before convert it", type(tokenized_datasets))
tokenized_datasets.set_format("torch")
print("type dataset", type(tokenized_datasets))
```

# Model

```python
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python
model
```

# Training

```python
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(
    output_dir='./results',
    weight_decay= 0.01,
    num_train_epochs = 3,
    per_gpu_eval_batch_size=8,
    per_device_eval_batch_size=3,
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    load_best_model_at_end=True
)
```

```python
from transformers import Trainer
import evaluate

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"].select(range(500)),
    eval_dataset=tokenized_datasets["test"].select(range(500)),
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

# Inference

- Methode 1

```python
pipeline  = pipeline("text-classification",
    model = "./bert_finetune" ,
                     tokenizer = model_dir)
sentence = "ce film est interessant!"
pipeline(sentence)
```

- Methode 2

```python
load_model = AutoModelForSequenceClassification.from_pretrained(model_dir)
load_tokenizer = AutoTokenizer.from_pretrained(model_dir)
print(load_model.config)
```

```python
input = load_tokenizer(sentence, return_tensors="pt")

outputs = load_model(**input)
# Convertir les logits en classe prédite
predictions = torch.argmax(outputs.logits, dim=-1)

print("Classe prédite :", predictions.item())
```

- With cuda

```python
import torch

# Préparer le texte pour l'inférence
text = "Ce film était incroyable !"
inputs = load_tokenizer(text, return_tensors="pt")

# Mettre le modèle sur GPU si disponible
device = "cuda" if torch.cuda.is_available() else "cpu"
load_model.to(device)
inputs = {k: v.to(device) for k, v in inputs.items()}

# Obtenir les prédictions
with torch.no_grad():
    outputs = load_model(**inputs)

# Convertir les logits en classe prédite
predictions = torch.argmax(outputs.logits, dim=-1)

print("Classe prédite :", predictions.item())  # Affiche la classe (ex. 0 ou 1)
```
